```python
# YOLOv11x 모델 불러오기
model = YOLO("yolo11x.pt")  # 로컬 경로 또는 Colab에 업로드된 경우 사용

# 객체 수 세기
object_counts = {}

# YOLO는 영상 분석 시, 프레임별 결과 리스트로 반환함
for result in results:
    for box in result.boxes:
        class_id = int(box.cls[0])  # 클래스 ID
        class_name = model.names[class_id]  # 클래스 이름

        # 클래스별 개수 세기
        if class_name in object_counts:
            object_counts[class_name] += 1
        else:
            object_counts[class_name] = 1

# 4. 결과 출력
print("🎯 감지된 객체 수:", object_counts)
```
